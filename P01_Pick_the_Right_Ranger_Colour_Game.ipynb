In this project, we are building an interactive game where the user competes against the computer in a color-matching challenge, inspired by the Power Rangers Dino Thunder theme. The user is prompted to select a color associated with a Dino Thunder Rangerâ€”such as Red Ranger, Blue Ranger, Yellow Ranger, Black Ranger, or White Rangerâ€”while the computer randomly selects one of these Ranger colors.

The goal is straightforward: if the user's chosen Ranger color matches the computer's randomly selected Ranger color, the user wins the round. Otherwise, the computer wins. After each round, the score is updated, with the user receiving 1 point for a win and the computer receiving 1 point for a loss, and vice versa.

Once the result of the round is declared, the computer asks the user if they would like to continue playing. If the user chooses to continue, the game resets for the next round. If the user decides to stop, the game concludes, and the final score is displayed.

This project utilizes fundamental Python programming concepts such as conditionals, loops, and random number generation, offering an engaging and thematic way to develop coding skills through game development.

### **Importing Necessary Libraries:**

```python
import random
```

### **Declare computer_points and player_points as variables; Show color choices to user to choose and enter the color:**

```python
# Print multiline instructions
print('''
Welcome to the Game!
\nIn the Pick the Right Ranger Color Game, choose a number from one to five to select your Ranger color.
\nMatch the computerâ€™s randomly chosen color to win the round!
''')

computer_points = 0
player_points = 0

while True:
    print("Red Ranger = 1\nBlue Ranger = 2\nYellow Ranger = 3\nBlack Ranger = 4\nWhite Ranger = 5")

    # Take the input from user and validate for valid number
    while True:
        try:
            ranger_choice = int(input("Itâ€™s Morphin Time! Choose a number (1-5) to select your Ranger and prepare for battle: "))
            if ranger_choice < 1 or ranger_choice > 5:
                print('Oops! Wrong Ranger number. Try again and pick a number between 1 and 5.')
            else:
                break
        except ValueError:
            print('Invalid input! Please enter a number between 1 and 5.')

    # Initialize ranger_color based on ranger_choice
    if ranger_choice == 1:
        ranger_color = 'Red Ranger'
    elif ranger_choice == 2:
        ranger_color = 'Blue Ranger'
    elif ranger_choice == 3:
        ranger_color = 'Yellow Ranger'
    elif ranger_choice == 4:
        ranger_color = 'Black Ranger'
    else:
        ranger_color = 'White Ranger'

    # Print user ranger choice
    print(f'User Ranger choice is: {ranger_color}')
    print('Now itâ€™s the computerâ€™s turn to choose a Ranger...')

    # Computer chooses randomly any number from 1 to 5
    computer_choice = random.randint(1, 5)

    # Initialize computer_color based on computer_choice
    if computer_choice == 1:
        computer_color = 'Red Ranger'
    elif computer_choice == 2:
        computer_color = 'Blue Ranger'
    elif computer_choice == 3:
        computer_color = 'Yellow Ranger'
    elif computer_choice == 4:
        computer_color = 'Black Ranger'
    else:
        computer_color = 'White Ranger'

    # Print the computer choice
    print(f'Computer Ranger choice is: {computer_color}')

    # Conditions for winning
    if ranger_color == computer_color:
        player_points += 1
        print('Player Points: ' + str(player_points))
        print('Computer Points: ' + str(computer_points))
    else:
        computer_points += 1
        print('Player Points: ' + str(player_points))
        print('Computer Points: ' + str(computer_points))

    # Ask if the user wants to continue and validate for 'Y' or 'N'
    while True:
        ans = input('Do you want to continue? (Y/N): ').strip().lower()
        if ans in ['y', 'n']:
            break
        else:
            print('Invalid input! Please enter Y to continue or N to stop.')

    # If user input 'n', exit the loop
    if ans == 'n':
        break
```

**Thanks for playing and the overall result of the game:**

```python
# After coming out of the loop, print the final result
if computer_points == player_points:
    print('Game is Tied')
elif computer_points < player_points:
    print('<== User wins ==>')
    print('\nPlayer is Victorious!')
else:
    print('<== Computer wins ==>')
    print('\nPlayer is Defeated!')

print('\nThanks for playing.')
```
